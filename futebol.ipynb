```python
!pip install streamlit
import requests
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import streamlit as st

url_artilheiros = "https://api.football-data.org/v4/competitions/PD/scorers"
url_partidas = "https://api.football-data.org/v4/competitions/PD/matches"
headers = {
    "X-Auth-Token": "72ad710724b54c519d11438290fe57d8"
}

response = requests.get(url_artilheiros, headers=headers)
response_partidas = requests.get(url_partidas, headers=headers)
data = response.json()
data_partidas = response_partidas.json()
```

```python
artilheiros = [art['player']['name'] for art in data['scorers']]
team = [art['team']['name'] for art in data['scorers']]
pga = [[art['playedMatches'], art['goals'], art['assists'] if art['assists'] is not None else 0] for art in data['scorers']]
time_casa = [partidas['homeTeam']['name'] for partidas in data_partidas['matches']]
time_fora = [partidas['awayTeam']['name'] for partidas in data_partidas['matches']]
gols_casa = [partidas['score']['fullTime']['home'] for partidas in data_partidas['matches']]
gols_fora = [partidas['score']['fullTime']['away'] for partidas in data_partidas['matches']]
print(time_casa)
print(time_fora)

df = pd.DataFrame({
    'time_casa': time_casa,
    'time_fora': time_fora,
    'gols_casa': gols_casa,
    'gols_fora': gols_fora
})

df['gols_casa'] = df['gols_casa'].fillna(0).astype(int)
df['gols_fora'] = df['gols_fora'].fillna(0).astype(int)

df.head(10)
df.to_csv('dados_partidas.csv', index=False)
```

```python
junto = list(zip(artilheiros, team, pga))
print(junto)
```

```python
dados_df = []
for nome, time, stats in junto:
    dados_df.append({
        'Artilheiro': nome,
        'Time': time,
        'Partidas': stats[0],
        'Gols': stats[1],
        'Assistencias': stats[2]
    })

df = pd.DataFrame(dados_df)

df.head(10)
df.to_csv('dados_artilheiros.csv', index=False)
```
